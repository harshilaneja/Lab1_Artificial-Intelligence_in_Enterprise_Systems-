```python
import pandas as pd


dataset_names = ['spotify_artist_data_2023.csv', 'spotify_data_12_20_2023.csv', 'spotify_features_data_2023.csv',
                 'spotify_tracks_data_2023.csv', 'spotify-albums_data_2023.csv']

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each dataset name and read its data into a DataFrame
for name in dataset_names:
    file_path = f'{name}'
    current_data = pd.read_csv(file_path)
    dfs.append(current_data)

# Concatenate the list of DataFrames into a single DataFrame
combined_data = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_data.to_csv('combined_dataset.csv', index=False)

print("Datasets have been successfully combined and saved.")
```

```python

missing_percentage = (df.isnull().sum() * 100 / len(df)).reset_index()
missing_percentage.columns = ['Column', 'Percentage']

# Plotting the percentage of missing values
plt.figure(figsize=(16, 5))
ax = sns.barplot(x='Column', y='Percentage', data=missing_percentage)
plt.xticks(rotation=90, fontsize=7)
plt.title("Percentage of Missing Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Percentage")
plt.show()
```

```python
# Univariate Analysis
for i, predictor in enumerate(df.drop(columns=['artist_genres', 'album_name', 'release_date','artists','label','track_id','track_name','mode'])):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='mode')
```
